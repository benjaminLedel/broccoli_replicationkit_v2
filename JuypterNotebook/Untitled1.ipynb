```python
import pandas as pd

data = pd.read_csv('model_data_SMARTSHARK_compress.csv')
data
```

```python
projects = data.project.unique()
projectVersionMap = {}
for project in projects:
    projectParts = project.split("_")
    groupName = projectParts[0] + "_" + projectParts[1]
    if groupName not in projectVersionMap:
        projectVersionMap[groupName] = []
    projectVersionMap[groupName].append(project)

projectVersionMap
```

```python
keys = ['smartshark_commons-codec', 'smartshark_commons-collections', 'smartshark_commons-compress','smartshark_commons-configuration','smartshark_commons-io','smartshark_commons-lang','smartshark_commons-math']
keylist = [projectVersionMap[key] for key in keys]
projectslist = [item for keylist in keylist for item in keylist]
print(projectslist)

subData2 = data[data['project'].isin(projectslist)]
subData2.to_csv('data_notTimeAware/Smartshark_train.csv')
```

```python



for key in projectVersionMap:
    subData = data[data['project'].isin(projectVersionMap[key])]
    subData.to_csv('data_notTimeAware/' + key +'.csv')
    subData2 = data[~data['project'].isin(projectVersionMap[key])]
    subData2.to_csv('data_notTimeAware/' + key +'_train.csv')
    
```

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
import xgboost as xgb

params = {}
num_trees=500
start_at = 'Previous_PDE'
found = True
for key in projectVersionMap:
    if start_at == key:
        found = True    
    if not found:
        continue
    
    print("Build Model " + key)
    train = pd.read_csv('data/' + key + '_train.csv')
    columns = train.drop(['project','bug','Result'], axis=1).columns
    gbm = xgb.train(params, xgb.DMatrix(
    train[columns], train['Result']), num_trees)
    gbm.save_model('model/' + key + '.model')
```

```python
import os
import pandas as pd
import xgboost as xgb
import pickle

rows_list = []
for root, dirs, files in os.walk("model"):
    for filename in files:
        if "pickle" in filename:
            #print(filename)
            loaded_model = pickle.load(open("model/" + filename, "rb"))
            e = loaded_model.get_score(importance_type='gain')
            rows_list.append(e)

df = pd.DataFrame(rows_list)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

index_sort = df.mean().sort_values(ascending=False).index
index_sort

sns.boxplot(data=df[index_sort])
# Label x-axis
plt.xlabel('Feature')
  
# labels y-axis
plt.ylabel('Relevance')
plt.xticks(rotation=90)
plt
```

```python
import os
import pandas as pd
import xgboost as xgb

rows_list = []
for root, dirs, files in os.walk("model"):
    for filename in files:
        if "smartshark" not in filename and ".json" in filename:
            #print(filename)
            gbm = xgb.Booster(model_file='model/' + filename)
            e = gbm.get_score(importance_type='gain')
            rows_list.append(e)

df = pd.DataFrame(rows_list)

df = df.rename(columns={'f0': "scoreFilename", 
                        "f1": "scoreBugReportSimple", 
                        "f2": "scoreClassNames" , 
                        "f3": "scoreMethodNames",
                        "f4": "scoreContent",
                       "f5": "scoreMethod",
                       "f6": "scorePath",
                       "f7": "scoreLOC",
                       "f8": "scoreBugreportsSummary",
                       "f9" : "scoreBugreportsDescription",
                       "f10": "scoreVersionHistory",
                       "f11": "scoreStacktrace"})

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

index_sort = df.mean().sort_values(ascending=False).index
index_sort

sns.boxplot(data=df[index_sort])
# Label x-axis
plt.xlabel('Feature')
  
# labels y-axis
plt.ylabel('Relevance')
plt.xticks(rotation=90)
plt.savefig("experiment_feature_frequency.png",bbox_inches='tight')
plt
```
